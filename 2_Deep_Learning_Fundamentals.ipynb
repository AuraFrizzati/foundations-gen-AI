## **2.2: What is a Perceptron**

A **perceptron** is an **essential component** in the world of AI, acting as a **binary classifier** capable of deciding whether data, like an image, belongs to one class or another.

It works by **adjusting its weighted inputs**—think of these like dials **fine-tuning** a radio signal—until it becomes better at predicting the right class for the data. This process is known as **learning**, and it shows us that even complex tasks start with small, simple steps.




image.png

### **Technical Terms Explained**:
- **Perceptron**: A **basic computational model** in machine learning that makes **decisions** by **weighing input data**. It's like a mini-decision maker that labels data as one thing or another.

- **Binary Classifier**: A type of system that **categorizes data** into **one of two groups**. Picture a light switch that can be flipped to either on or off.

- **Vector of Numbers**: A **sequence of numbers arranged in order**, which together represent one piece of data.

- **Activation Function**: A **mathematical equation** that **decides whether the perceptron's calculated sum from the inputs is enough to trigger a positive or negative output**.

Given an input (a vector of number), the perceptron decides if it belongs to one of 2 classes. In essence, a perceptron is a model of a single neuronal unit that can be used for binary classification.




### **How does a perceptron learn?**

- During the process called **learning**, the perceptron receives **multiple input signals** (e.g. images of cats and dogs). **Each input vector passes through the entire perceptron** and t**he output classification is compared with the actual answer**.

- **Before the weights are learned**, chances are the **predictions are often incorrect**. The perceptron then **nudges the weights** so to increase the likelihood to obtain the correct answer. After a lot of nudging, the perceptron should improve its performance.

- A single perceptron will probably not be able to learn the required classification task too well, but when you **combine the power of many perceptrons together**, it will be possible.


### **The Perceptron activation function**

Historically, the **perceptron** employed a **step function** as its activation function, making it a **binary output system** that could take only one of two values (0 or 1). Technically it can also take the value 0.5 but only for a single point (and this is not relevant for the current discussion)

In today's **advanced neural network architectures**, activation functions have evolved. For example, the **ReLu** (**Rectified Linear Unit**) **activation function** is a popular choice in the field. A ReLu can **return any non-zero number**, allowing for a **broader range of output values** and greater nuance in the learning process. A ReLu activation function doesn't jump directly from 0 to 1 but climbs up gradually and actually keeps on climbing after 1...

## **2.3: The Multi-Layer Perceptron**

The multi-layer perceptron is a powerful tool in the world of machine learning, capable of making smart decisions by mimicking the way our brain's neurons work. This amazing system can learn from its experiences, growing smarter over time as it **processes information through layers**, and eventually, it can predict answers with astonishing accuracy!

### **Technical Terms Explained**:
- **Multi-Layer Perceptron** (**MLP**): A type of **artificial neural network** that has **multiple layers of nodes**, each layer learning to recognize increasingly complex features of the input data.

- **Input Layer**: The first layer in an MLP where the **raw data** is **initially received**.

- **Output Layer**: The **last layer** in an MLP that **produces the final result or prediction** of the network.

- **Hidden Layers**: Layers between the input and output that **perform complex data transformations**.

MLP is a type of artificial neural network composed of multiple layers of nodes or neurones. Each **node/neuron** is a **perceptron** and each performs a simple computation that passes the result to the next layer.

The MLP power comes from its ability to learn from experience.

- **Each perceptron in the hidden layer is connected to each of the inputs**, and each of these **connections** is **associated with a weight**. During **training**, the **weights are adjusted** so to **reduce the number of errors** the model makes overall.

- **Each of the neurons in a subsequent layer are connected to each of the neurones in its preceding layer**. Also their connections' weights are adjusted during training.

- In the **final layer** we usually have a **number of neurones equal to the number of classes to predict**. One neuron will have the highest value and that's the one whose class we choose

### **Understanding Gradient Descent**

**Gradient descent** is a vital **optimisation algorithm** used to **fine tune parameters** in deep learning models. Its purpose is to **minimise** a function referred to as **cost ot objective function**.

In a mathematical context, the **gradient** represents the **direction of the steepest increase of a function**. **Descending** the **gradient** implies **moving opposite to this direction** (i.e. downhill). During each **iteration**, the **algorithm adjusts the weights** of the model with respect to an amount we define called the **learning rate**, which determines the step size during each iteration. If the **rate** is **too large**, we **risk overshooting the minimum**, if it is **too small** a rate, it may lead to **protracted convergence times**

image.png

As **gradient descent progresses**, it **iteratively refines the parameters** to **minimise a cost**, leading to an **optimised model**.

### **Using Backpropagation to Train the Neural Network**

**Backpropagation** is a shorthand for **backward propagation of errors** and is a **standard method** for **training** neural networks. It **adjusts the network's weights** based on the **error** produced in the **previous iteration**.

The process commences with a **forward pass**, where the **input traverses through the network layers** to produce an **output**, which is then **compared to the expected results**, producing an **error value**.

The crux of backpropagation is to **distribute this error backward through the network**. The **error** is p**ropagated to each neuron**, attributing a blame for a portion of the total error. For **each weight in the network**, a **gradient is computed**, which indicates **how much the error will change if the error is adjusted by a small amount**. **Weights** are then **updated** using the **gradient** and a **learning rate**.

By **iteratively performing forward passes and backpropagations**, the **network weights are honed**, enabling it to make more accurate predictions.

### **Testing your model**

After training your model, you need to **test** it on a **dataset different from the one used during training**. Sometimes this is called the **hold-out dataset** or **test dataset**. You want to see how well your model performs on data it has not seen. This should indicate how well the model perform when it starts making predictions in the real world.

### **Step 1: Load the dataset**
We obtain the MNIST dataset using scikit-learn.

```python
# Load MNIST using sklearn.datasets.fetch_openml
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import fetch_openml
```

```python
# Load data from https://www.openml.org/d/554
X, y = fetch_openml("mnist_784", # 28x28 = 784 features

                    # This specifies that we want to download version 1
                    # of the dataset
                    version=1,

                    # This tells the function to return the data (X) and the
                    # labels (y) separately.
                    return_X_y=True,

                    # nstructs the function to automatically determine
                    # the best parser to use based on the downloaded
                    # data format
                    parser="auto"
                    )
```

```python
# Split into train and test
X_train, X_test = X[:60000], X[60000:]
y_train, y_test = y[:60000], y[60000:]
print(f"Training set size: {len(X_train)}")
print(f"Test set size: {len(X_test)}")
```

```python
# Convert to numpy arrays and scale for the model
## Normalizing the pixel values helps the neural network learn more effectively.
X_train = np.array(X_train) / 255 # scale from 0-255 to 0-1
X_test = np.array(X_test) / 255
y_train = np.array(y_train, dtype=np.int8)
y_test = np.array(y_test, dtype=np.int8)
```

```python
# Show the first 3 images
plt.figure(figsize=(20, 4))
for index, (image, label) in enumerate(zip(X_train[0:3], y_train[0:3])):
    plt.subplot(1, 3, index + 1)
    plt.imshow(np.reshape(image, (28, 28)), cmap=plt.cm.gray)
    plt.title("Label: %s\n" % label, fontsize=20)
```

### **Step 2: Create and train a Multi-Layer Perceptron**
We will use `sklearn.neural_network` to build a Multi-Layer Perceptron in a single command. We'll train it using a second command!

```python
# Train an MLP classifier using sklearn.neural_network.MLPClassifier
from sklearn.neural_network import MLPClassifier

# Create an MLPClassifier object
mlp = MLPClassifier(
    hidden_layer_sizes=(50,),
    max_iter=10, # 10 epochs
    alpha=1e-4,  # Strength of the L2 regularization term
    solver="sgd",# stochastic gradient descent
    verbose=10,
    random_state=1, # random seed for reproducibility
    learning_rate_init=0.1, # The initial learning rate used. It controls the step-size in updating the weights.
)


# Train the MLPClassifier
mlp.fit(X_train, y_train)
```

### **Step 3: Evaluate the model on a hold-out dataset**
We will assess the performance of the model on data that was not used for training. In this case, we had separated out data into training and test splits, so we will show the accuracy for each.

```python
# Show the accuracy on the training and test sets

print(f"Training set score: {mlp.score(X_train, y_train)}")
print(f"Test set score: {mlp.score(X_test, y_test)}")
```

Lastly, even though metrics are important to look it, it's always a good idea to look directly at your data to catch any problems or bugs you might have missed.

```python
# Show the images, predictions, and original labels for 10 images

# Get the predictions for the test dataset
predictions = mlp.predict(X_test)

# Show the predictions in a grid
plt.figure(figsize=(8, 4))

for index, (image, prediction, label) in enumerate(
    zip(X_test[0:10], predictions[0:10], y_test[0:10])
):
    plt.subplot(2, 5, index + 1)
    plt.imshow(np.reshape(image, (28, 28)), cmap=plt.cm.gray)

    # Green if correct, red if incorrect
    fontcolor = "g" if prediction == label else "r"
    plt.title(
        "Prediction: %i\n Label: %i" % (prediction, label), fontsize=10, color=fontcolor
    )

    plt.axis("off")  # hide axes
```

### **PyTorch MLP Class**

Let's create a one hidden layer MLP using PyTorch

```python
# torch.nn module contains the building blocks for creating
# neural networks in PyTorch
import torch.nn as nn

class MLP(nn.Module): # MLP inherits from nn.Module (base nn class in PyTorch)
  ## The class instantiator let's the user specify the input size
  def __init__(self,input_size):

    # This line initialises the parent class, ensuring all the necessary
    # setup for a PyTorch module is done
    super(MLP,self).__init__()

    # We use the nn.Linear method to create neural net layers
    ## 64 is the number of nodes in the hidden layer
    ## input_size x 64 = number of weights in the hidden layer
    self.hidden_layer = nn.Linear(input_size,64)
    ## 64 x 2 = 128, number of weights in the output layer
    self.output_layer = nn.Linear(64,2)
    ## activation function
    self.activation = nn.ReLU()

  # We then define how the components are actually connected to each other
  # in the forward method. x represents the input data.
  def forward(self,x):
    ## The result first is passed through the activation function
    x = self.activation(self.hidden_layer(x))
    ## Finally, the result is passed through the output layer
    return self.output_layer(x)

```

We can now **instantiate** and **print the MLP** to review the sizes of our layers:

```python
model = MLP(input_size = 10)
print(model)
```

We can then **test out** the **forward method the MLP model** by passing in a vector of length 10. As expected, we receive a vector of length 2.

```python
model.forward(x = torch.rand(10))
```

- `nn.Module`: fundamental PyTorch **building block**
- `nn.Linear`: **fully connected layer**
- `nn.ReLu`: **non-linear activation function**
- `forward` method: it defines **forward computation**

In PyTorch, `nn.Linear` is a **module** that creates a **linear transformation** (a simple, but important, mathematical operation) within a neural network.

Think of it like this:
1. **Input**: You have some input data, which in the context of neural networks is often represented as a **vector** (a list of numbers).
2. **Weights** and **Biases**: `nn.Linear` introduces a set of weights and biases. **Weights** determine the **importance of each input feature**, while **biases** provide an **offset**.
3. **Linear Transformation**: The **input** data is **multiplied by the weights**, and **the biases are added**. This process is a **linear transformation**, and it's a core operation in many neural network layers.
4. **Output**: The **result** of this **linear transformation** is the **output** of the `nn.Linear` module, which is then **typically passed to the next layer** in the network or used to make predictions.

## **2.9: PyTorch Loss Functions**
PyTorch loss functions are essential **tools** that help in **improving the accuracy of a model by measuring errors**. These functions come in **different forms** to **tackle various problems**, like **deciding between categories** (**classification**) or **predicting values** (**regression**). Understanding and using these functions correctly is key to making smart, effective models that do a great job at the tasks they're designed for!

### **Technical Terms Explained**:
- **Loss functions**: They **measure how well a model is performing** by calculating the **difference** between the model's **predictions** and the **actual results**.

- **Cross entropy loss**: This is a measure used when a model needs to choose between **categories** (like whether an image shows a cat or a dog), and it shows **how well the model's predictions align with the actual categories**.

- **Mean squared error**: This shows the **average of the squares of the differences between predicted numbers** (like a predicted price) **and the actual numbers**. It's often used for **predicting continuous values** rather than categories.

**Loss functions** play a pivotal role in **guiding model optimisation**. They **quantify** the **discrepancy** between the **predicted output** and the **actual target values**. By **minimising** this **quantified error**, a model can be trained to produce more accurate results.

**Pythorch** offers a **coprehensive suite of loss functions** through its `torch.nn` module. Two commonly enployed ones are:

- **Cross-Entropy Loss**: suited for **classification tasks**, particularly useful when the **classes** are **mutually exclusive**
- **Mean Squared Error** (**MSE**): used primarily for **regression**. It calculates the **average** of the **square differences** between the **predicted** and **target values**

For both **measures of loss**, a **lower value** indicates a **model** that **better approximates** the true **data**

### **Cross-Entropy Loss in PyTorch**

We **define** a **target tensor** that we will use for **evaluating** our **predictions**

```python
## This line imports the PyTorch library, which is the core library
## for deep learning in Python. It provides tools for creating and
## manipulating tensors, defining neural networks and performing
## computations on them
import torch

## This line imports the nn module from PyTorch. It contains various
## building blocks for creating neural networks, such as layers,
## activation functions and loss functions.
import torch.nn as nn

## This creates an object of the CrossEntropyLoss class, which will be used
## to calculate the loss during training
loss_function = nn.CrossEntropyLoss()

# Our dataset contains a single image of a dog, where
# cat = 0 and dog = 1 (corresponding to index 0 and 1)

## This creates a PyTorch tensor with a single element, which is the
## value 1. In this case, it indicates that the image is classified
## as a "dog"
target_tensor = torch.tensor([1])
target_tensor
```

Let's see what the cross-entropy looks like in two separate cases:

1. Our predicted tensor is [2, 5]. Since the second number is larger, this means the model predicts that the input is more likely a dog (index 1 is larger)

```python
# Prediction: Most likely a dog (index 1 is higher)
# Note that the values do not need to sum to 1
## This part simulates a prediction from our model and then calculates
## the loss using the previously defined loss function

## This creates a tensor representing the model's prediction. It's a 2D tensor
## because the model is likely outputting probabilities or scores for each class
## (cat and dog). The higher value at index 1 indicates that the model believes
## the input is more likely a dog
predicted_tensor = torch.tensor([[2.0, 5.0]])

## This line is where the cross-entropy loss is calculated. The loss_function
## object is called and compares the model's prediction to the true label and
## returns a loss value
loss_value = loss_function(predicted_tensor, target_tensor)
loss_value
```

2. Prediction: Slightly more likely a cat (index 0 is higher)

```python
predicted_tensor = torch.tensor([[1.5, 1.1]])
loss_value = loss_function(predicted_tensor, target_tensor)
loss_value
```

## **2.10: PyTorch Optimisers**

PyTorch optimisers are important tools that help **improve how a neural network learns from data** by **adjusting the model's parameters**. By using these optimizers, like stochastic gradient descent (SGD) with momentum or Adam, we can quickly get started learning!

### **Technical Terms Explained:**
- **Gradients**: **Directions** and **amounts** by which a **function increases most**. The **parameters** can be **changed in a direction opposite to the gradient of the loss function** in order to reduce the loss.

- **Learning Rate**: This **hyperparameter** specifies **how big the steps** are when **adjusting the neural network's settings** during **training**. Too big, and you might skip over the best setting; too small, and it'll take a very long time to get there.

- **Momentum**: A **technique** that helps **accelerate the optimizer in the right direction** and **dampens oscillations**.

PyTorch Optimisers are essential components in the training process of neural networks. Their primary function is to **adjust the parameters** of a **model** in response to the **computed gradients** aiming to **minimise** the **loss function**. This is crucial to improve the model's accuracy and performance on unseen data

To alleviate this, techniques such as incorporating **momentum** are often introduced. Momentum **accumulates past gradients to smooth updates** (**optimisation trajectories**), **reducing oscillations**

```python
import torch
import torch.nn as nn

## the PyTorch module optim contains various optimisation algorithms
## used to update the parameters of your deep learning model during
## training
import torch.optim as optim

# Define a simple model
class SimpleModel(nn.Module):
    def __init__(self):
        super(SimpleModel, self).__init__()

        ## We create a single linear layer that takes an onput of
        ## size 10 and produces an output of size 1
        self.linear = nn.Linear(10, 1)  # Input size 10, output size 1

    ## This defines the forward pass of the model, specifying
    ## how the input data (x) flows through the layers to produce
    ## the output. In this case, the input is simply passed through the
    ## linear layer
    def forward(self, x):
        return self.linear(x)

# Create an instance of the model
model = SimpleModel()

# momentum=0.9 smoothes out updates and can help training
## This line creates an optimiser using the SGD algorithm. SGD is a
## popular optimisation algorithm that iteratively adjusts the model's
## parameters to minimise errors
optimizer = optim.SGD(

    ## These are the adjustable parameters (weights and biases) within
    ## your neural network model. The optimiser needs access to these
    ## parameters to update them during training
    model.parameters(),

    ## This sets the Learning Rate, which controls the size of the
    ## steps taken by the optimiser when updating a model's parameter
    lr=0.01,

    ## Momentum helps the optimiser to continue moving in the same
    ## direction even if the gradient (the direction of the steepest descent)
    ## changes abruptly. It acts like a smoothing factor, preventing
    ## the optimiser from getting stuck in local minima. This often helps
    ## training converge faster and more reliably
    momentum=0.9
    )
```

### **Why gradients might change abruptly during training?**
1. **Noisy Data**:
Real-world datasets often contain **noise** or **outliers**. These noisy data points can cause sudden and large changes in the loss function, leading to abrupt gradient changes. As the optimiser processes these noisy data points, it might make a sharp turn in its parameter updates, trying to fit the noise, which can be detrimental to overall learning.
2. **Non-Convex Loss Landscapes**:
Deep learning models often have complex and non-convex loss landscapes. These landscapes have **numerous local minima** and **saddle points** where the **gradient can be zero** or **change direction significantly**.
During training, the optimiser can get stuck in a flat region (a plateau) or jump between different local minima, causing the gradient to change abruptly as it navigates these challenging areas.
3. **Mini-Batch Gradient Descent**:
In mini-batch gradient descent, the **gradient** is **calculated using only a small subset** (a **mini-batch**) of the **training data**. This introduces a **level of stochasticity** (randomness) into the **gradient calculation**. As the **mini-batches change**, the **gradient** calculated from them can **vary significantly**, resulting in abrupt changes in the direction of parameter updates.


4. **Interactions between Parameters**: The **parameters** of a deep learning model are **highly interconnected**. Changes in one parameter can influence the gradients of other parameters in unexpected ways. During training, as the optimiser updates parameters, these **interactions** can create **complex dynamics** that lead to **sudden** and **large changes** in the **gradient**.
5. **Sharp Transitions in Data**: If the **training data** contains **sharp transitions** or **discontinuities**, the model might struggle to smoothly learn these changes.
This can result in the optimiser making large updates to the parameters in an attempt to fit these abrupt transitions, leading to abrupt gradient changes.
6. **Learning Rate Issues**: A learning rate that is **too high** can cause the optimiser to **overshoot** the **optimal parameter values** and **oscillate around the minima**, resulting in abrupt gradient changes. A **learning rate** that is **too low** can make the **optimizer too sensitive to noise and local minima**, potentially leading to abrupt changes in the gradient as well.

### **An example `Dataset`**

In **PyTorch**, **creating a custom dataset** involves **subclassing** the **`Dataset` class** and **implementing** the `__getitem__` and `__len__` (length) **methods**.

For example, let's define a dataset for returning pairs of consecutive numbers and their product:

```python
## The Dataset class in PyTorch is the foundation for creating
## custom datasets
from torch.utils.data import Dataset

# Create a toy dataset
class NumberProductDataset(Dataset):
  def __init__(self, data_range=(1,10)):
    self.numbers = list(range(data_range[0], data_range[1]))

  # The __getitem__ method provides indexed access to the
  # pairs ((num1,num2), num1*num2)
  ## This method allows you to access data samples using indexing
  def __getitem__(self, index):
    number1 = self.numbers[index]
    number2 = self.numbers[index] + 1
    return(number1, number2), number1 * number2

  # The __len__ method returns the total number of pairs
  def __len__(self):
    return len(self.numbers)

# Instantiate the dataset
dataset = NumberProductDataset(data_range=(0,11))

# Access a data sample
data_sample = dataset[3]
print(data_sample)
print(dataset)
len(dataset)
```

## **2.12: PyTorch Training Loops**

A **PyTorch training loop** is an essential part of building a neural network model, which helps us teach the computer how to make predictions or decisions based on data. By using this loop, we **gradually improve our model's accuracy** through a process of **learning** from its mistakes and adjusting.

### **Technical Terms Explained**:
- **Training Loop**: The **cycle** that a neural network goes through many times to **learn from the data** by **making predictions**, **checking errors**, and **improving itself**.

- **Batches**: Batches are **small**, **evenly divided parts of data** that the AI looks at and learns from each step of the way.

- **Epochs**: A **complete pass through the entire training dataset**. The more epochs, the more the computer goes over the material to learn.

- **Loss functions**: They **measure how well a model is performing** by calculating the **difference** between the **model's predictions** and the **actual results**.

- **Optimizer**: Part of the neural network's brain that **makes decisions on how to change the network to get better** at its job.

Let's bring together all the components of PyTorch and show a full-fledged training loop. For this examples, we will create a simple dataset consisting of pairs of numbers and their sums

### [1] Create a Number Sum Dataset

This dataset has two features—a pair of numbers—and a target value—the sum of those two numbers.

Note that **this is not actually a good use of deep learning**. At the end of our training loop, the model still doesn't know how to add 3 + 7! The idea here is to use a simple example so it's easy to evaluate the model's performance.

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

class NumberSumDataset(Dataset):

  def __init__(self, data_range=(1,10)):
    self.numbers = list(range(data_range[0], data_range[1]))

  def __getitem__(self, index):
    ## note: len(self.numbers) here doesn't rely on the
    ## special method __len__ define below but it is still
    ## len(self.numbers). In the example for
    ## NumberSumDataset(data_range=(1, 100)), it is 99

    ## Slower loop through all the numbers
    number1 = float(self.numbers[index // len(self.numbers)])

    ## Quicker loop through all the numbers
    number2 = float(self.numbers[index % len(self.numbers)])
    return torch.tensor([number1, number2]), torch.tensor([number1 + number2])

  def __len__(self):
    return len(self.numbers)**2

dataset = NumberSumDataset(data_range=(1, 100))

for i in range(5):
    print(dataset[i])

print(len(dataset)) ## 99^2
print(dataset[99])
print(dataset[100])
print(dataset[197])
print(dataset[198])
```

### [2] Define a simple Model



```python
class MLP(nn.Module):
  def __init__(self,input_size):
    super(MLP,self).__init__()
    self.hidden_layer = nn.Linear(input_size, 128)
    self.output_layer = nn.Linear(128, 1)
    self.activation = nn.ReLU()

  def forward(self, x):
    x = self.activation(self.hidden_layer(x))
    return self.output_layer(x)
```

### [3] Instantiate the components needed for training

**Training loops** vary in complexity but they all need these **core components**:

```python
dataset = NumberSumDataset(data_range=(0,100))
dataloader = DataLoader(dataset, batch_size = 100, shuffle = True)
model = MLP(input_size = 2)
loss_function = nn.MSELoss()
optimizer = optim.Adam(model.parameters(),lr=0.001)
```

### [4] Create a Training Loop

```python
## Outer loop: Iterate over Epochs (compleye passes through
## the entire dataset)
for epoch in range(10):
  ## loss initialisation
  total_loss = 0.0
  ## Inner loop: Iterate over the batches
  for number_pairs, sums in dataloader:
    predictions = model(number_pairs) # compute the model's output
    loss = loss_function(predictions,sums) # compute the loss

    ## This step initiates backpropagation, a process that calculates
    ## the gradients of the loss with respect to the model's parameters.
    ## Gradients indicate the direction and magnitude of adjustments
    ## needed to improve the model's predictions
    loss.backward() # perform backpropagation

    ## This line instructs the optimizer to update the model's parameters
    ## based on the calculated gradients. The optimizer is an algorithm
    ## that determines how to adjust the parameters to minimise the loss
    optimizer.step() # update the parameters

    ## After each parameter update, this line resets the gradients
    ## to zero. This is necessary to prevent gradients from accumulating
    ## across batches, which could lead to incorrect updates
    optimizer.zero_grad() # zero the gradients

    total_loss +=loss.item() # add the loss for all batches (current epoch)

  # Print the loss for this epoch
  print(f"Epoch {epoch}: Sum of Batch Losses = {total_loss:.5f}")
```

Notice that the loss decreases, meaning that the model has started to understand the data and is able to predict the outputs given the inputs

### **Softmax function**

https://www.geeksforgeeks.org/the-role-of-softmax-in-neural-networks-detailed-explanation-and-applications/

- Softmax function is a **mathematical function** that **converts** a **vector of raw prediction scores** (often called **logits**) from the neural network into **probabilities**. These probabilities are **distributed across different classes such that their sum equals 1**.
- Essentially, Softmax helps in **transforming output values** into a **format** that can be i**nterpreted as probabilities**, which makes it suitable for classification tasks.
- In a **multi-class classification neural network**, the **final layer outputs a set of values**, **each corresponding to a different class**. These values, before Softmax is applied, can be **any real numbers**, and may not provide meaningful information directly. The Softmax function processes these values into **probabilities**, which indicate the likelihood of each class being the correct one.

- Softmax gained prominence with the rise of **deep learning**, particularly in models such as **multilayer perceptrons** (**MLPs**) and **convolutional neural networks** (**CNNs**), where it is typically **applied to the final output layer in classification tasks**.

$$ Softmax(z_i) = \frac{e^{z_i}}{\sum^{K}_{j= 1}e^{z_j}} $$


- $z_i$ is the logit (the output of the previous layer in the network) for the $i^{th}$ class
- $K$ us the number of classes
- $e^{z_i}$ is the exponential of the logit
- $\sum^{K}_{j= 1}e^{z_j}$ is the sum of the exponentials across all classes
